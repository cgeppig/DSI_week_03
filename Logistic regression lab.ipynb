```python
evergreen['alchemy_category_score'].head()
```

```python
def bleach(string):
    temp = ""
    string = str(string)
    for i in string:
        if i in ["1","2","3","4","5","6","7","8","9","0", "."]:
            temp += i
    if len(temp) > 0:
        return float(temp)
```

```python
evergreen['alchemy_category_score'] = evergreen['alchemy_category_score'].apply(bleach)
```
