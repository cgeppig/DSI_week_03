```python


data = {}

counter = 0

for divtag in soup('div', {'class': 'zsg-photo-card-caption'}):
    data[counter] = {}
    for house_price in divtag('span', {'class': 'zsg-photo-card-price'}):
        data[counter]['price'] = house_price.string
    for address in divtag('span', {'class': 'zsg-photo-card-address'}):
        data[counter]['address'] = address.string
    for broker in divtag('span', {'class': 'zsg-photo-card-broker-name'}):
        data[counter]['broker'] = broker.string
    for house_info in divtag('span', {'class': 'zsg-photo-card-info'}):
        data[counter]['house_info'] = house_info.get_text()
    counter += 1
    
data
```

```python
pd.DataFrame(data)  ## columns and rows are reversed
```

```python
df = pd.DataFrame(data).T  ## .T is 'transpose'
```

```python
df.loc[0, 'house_info']
```

```python
df['bedrooms'] = df['house_info'].apply(lambda x: x.split(u'\xb7')[0])
df['bathrooms'] = df['house_info'].apply(lambda x: x.split(u'\xb7')[1])
df['square feet'] = df['house_info'].apply(lambda x: x.split(u'\xb7')[2])
```

```python
def bleach(string):
    temp = ""
    string = str(string)
    for i in string:
        if i in ["1","2","3","4","5","6","7","8","9","0"]:
            temp += i
    if len(temp) > 0:
        return float(temp)
```

```python
df.head()
```

```python
df['price_numeric'] = df['price'].apply(lambda x: ''.join(ch for ch in x if ch.isdigit()))
```

```python
for column in ['bedrooms', 'bathrooms', 'square feet', 'price', 'price_numeric']:
    df[column] = df[column].apply(bleach)
```

```python
right_table=soupwiki.find('table', class_='wikitable sortable plainrowheaders')
```

```python
print right_table.prettify()
```

```python
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
G=[]
for row in right_table.findAll("tr"):
    cells = row.findAll('td')
    states=row.findAll('th') #To store second column data
    if len(cells)==6: #Only extract table body not heading
        A.append(cells[0].find(text=True))
        B.append(states[0].find(text=True))
        C.append(cells[1].find(text=True))
        D.append(cells[2].find(text=True))
        E.append(cells[3].find(text=True))
        F.append(cells[4].find(text=True))
        G.append(cells[5].find(text=True))
```

```python
E
```

```python
df=pd.DataFrame(A,columns=['Number'])
df['State/UT']=B
df['Admin_Capital']=C
df['Legislative_Capital']=D
df['Judiciary_Capital']=E
df['Year_Capital']=F
df['Former_Capital']=G
df
```
